```python
# independent and dependent features

x = df.iloc[:,0:5]  # leaving last coloumn
y = df.iloc[:,4]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
from sklearn.model_selection import GridSearchCV

parameter = {'penalty':['l1','l2','elasticnet'],
            'C': [1,2,3,4,5,6,10,20,30,40,50],
            'max_iter':[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier, 
                                   param_grid = parameter,
                                   scoring = 'accuracy',
                                   cv = 5)
```

```python
classifier_regressor.fit(x_train, y_train)
```

```python
print(classifier_regressor.best_params)
```

```python
print(classifier_regressor.best_score)
```

```python
# Predection 

classifier_regressor.predict(x_test)
```

```python
y_pred = classifier_regressor.predict(x_test)

y_pred
```

```python
# Accuracy Score

from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
